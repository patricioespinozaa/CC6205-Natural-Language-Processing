### Creaci贸n b谩sica de arreglos

Creando e imprimiendo arreglos de numpy (ndarray):
- las matrices unidimensionales se imprimen luego como filas,
- las bidimensionales como matrices y tridimensionales como listas de matrices.

````
# Crear un arreglo a partir de una lista de listas
print("Arreglo a partir de una lista de listas:")
print(np.array([[1., 0., 0.], [0., 1., 2.]]), "\n")

# Crear un arreglo de ceros
print("Arreglo de ceros:")
print(np.zeros(10), "\n")

# Crear un arreglo de ceros multidimensional
a = np.zeros((2, 3, 4))
print("Arreglo multidimensional de ceros:")
print(a, "\n")

# Crear un arreglo de unos
print("Arreglo de unos:")
print(np.ones(10), "\n")

# Crear un arreglo de unos multidimensional con tipo de datos especificado
b = np.ones((2, 3, 4), dtype='int')
print("Arreglo multidimensional de unos:")
print(b, "\n")

# Crear un arreglo de valores aleatorios entre 0 y 1
print("Arreglo de valores aleatorios:")
print(np.random.rand(5), "\n")

# Crear un arreglo multidimensional de valores aleatorios entre 0 y 1
c = np.random.rand(2, 3, 4)
print("Arreglo multidimensional de valores aleatorios:")
print(c, "\n")
````

Los atributos m谩s importantes de un objeto ndarray son:

````
# N煤mero de dimensiones del arreglo
print("N煤mero de dimensiones del arreglo:", c.ndim)

# Forma del arreglo (tama帽o en cada dimensi贸n)
print("Forma del arreglo (tama帽o en cada dimensi贸n):", c.shape)

# N煤mero total de elementos en el arreglo
print("N煤mero total de elementos en el arreglo:", c.size)

# Tipo de datos de los elementos del arreglo
print("Tipo de datos de los elementos del arreglo:", c.dtype)
````

````
# Cambiando las dimensiones de un array (ravel, reshape, resize)

# Utilizar 'ravel' para aplanar el arreglo a una dimensi贸n
d = c.ravel()
print("Arreglo aplanado:", d, "\n")
print("Forma del arreglo aplanado:", d.shape, "\n")
print("Forma del arreglo original:", c.shape, "\n")

# Utilizar 'reshape' para crear un nuevo arreglo con otra distribuci贸n de las dimensiones (no modifica el original)
d = c.reshape((2, 12))
print("Nuevo arreglo con otra distribuci贸n de las dimensiones:", d, "\n")
print("Forma del nuevo arreglo:", d.shape, "\n")

# Utilizar 'T' para crear un nuevo arreglo con la transpuesta
e = d.T
print("Nuevo arreglo con la transpuesta:", e, "\n")
print("Forma del nuevo arreglo con la transpuesta:", e.shape, "\n")

# Utilizar 'resize' para cambiar las dimensiones en el propio arreglo (modifica el original)
x = d.resize(2, 12)
print("Forma del arreglo modificado:", d.shape, "\n")
print("Valor devuelto por resize (None indica que el arreglo fue modificado in-place):", x)
````

````
# Crear un arreglo de unos con dimensiones similares a otro arreglo
b = np.ones_like(a)
print("Arreglo de unos con dimensiones similares a 'a':")
print(b, "\n")
print("Forma del arreglo 'b':", b.shape, "\n")

# Crear un arreglo vac铆o con dimensiones similares a otro arreglo
c = np.empty(a.shape)
print("Arreglo vac铆o con dimensiones similares a 'a':")
print(c, "\n")
print("Forma del arreglo 'c':", c.shape, "\n")
````

### Broadcasting

Muchas operaciones en NumPy soportan [Broadcasting Semantics](https://docs.scipy.org/doc/numpy/user/basics.broadcasting.html#module-numpy.doc.broadcasting).

Si una operaci贸n NumPy admite broadcast, sus argumentos (ndarrays) se pueden expandir autom谩ticamente para que sean del mismo tama帽o (**sin hacer copias de los datos**). *No es m谩s que una forma de hacer operaciones punto a punto sobre ndarrays que no necesariamente tienen las mismas dimensiones.*

Dos `ndarray` son "broadcastable" si:
1. Cada array tiene al menos una dimensi贸n.
2. Al iterar sobre los tama帽os de las dimensiones (comenzando en la dimensi贸n final), sus tama帽os deben ser iguales, uno de ellos es `1` o uno de ellos no existe.

Si dos `ndarray` son "broadcastable" las dimensiones del `ndarray` resultado se determina mediante:
1. Si el n煤mero de dimensiones no es igual, se antepone `1` a las dimensiones del `ndarray` con menos dimensiones para que tengan la misma longitud.
2. Luego, para cada tama帽o de dimensi贸n, el tama帽o de dimensi贸n resultante es el m谩ximo de los tama帽os en esa dimensi贸n.

````
# Definir los arreglos con np.random.randint
a = np.random.randint(0, 10, size=(2, 1))
b = np.random.randint(0, 10, size=(1, 8))

# Mostrar los arreglos
print("Arreglo 'a':")
print(a)
print("\nArreglo 'b':")
print(b)

# Realizar la suma y obtener la forma resultante
result = a + b

# Mostrar el resultado de la suma y su forma
print("\nResultado de la suma:")
print(result)
print("\nForma del resultado de la suma:", result.shape)
````

El broadcasting se aplica para extender autom谩ticamente las dimensiones de ambos arreglos de manera que sean compatibles para la operaci贸n de suma. NumPy agrega dimensiones adicionales con tama帽o 1 a los arreglos que lo necesiten, permitiendo as铆 la realizaci贸n de operaciones elemento a elemento. En este caso, el resultado es un arreglo con forma (2, 8), donde cada elemento es la suma del correspondiente elemento en 'a' y 'b'.

### Tokenization

Importar el tokenizador de palabras recomendado por NLTK (actualmente ```nltk.tokenize.treebank.TreebankWordTokenizer``` junto con ```nltk.tokenize.punkt.PunktSentenceTokenizer``` para el idioma especificado)


````
# Importar tokenizador
from nltk import word_tokenize

# Frase de ejemplo
s = "Hola como est谩n, esta es una frase en #espa帽ol :)"

# Tokenizar la frase en espa帽ol
tokens = word_tokenize(s, language="spanish")

# Mostrar los tokens resultantes
tokens
````

### Stemming y Lemmatization

Importar stemmer ```SnowballStemmer``` desde NLTK. SnowballStemmer soporta los lenguajes: Arabic, Danish, Dutch, English, Finnish, French, German, Hungarian, Italian, Norwegian, Portuguese, Romanian, Russian, Spanish and Swedish.

````
# Importar stemmer
from nltk.stem import SnowballStemmer

# Crear una instancia del SnowballStemmer para espa帽ol
stemmer = SnowballStemmer("spanish")

# Aplicar stemming a diferentes formas de la palabra "correr"
print(stemmer.stem("corriendo"))
print(stemmer.stem("correr"))
print(stemmer.stem("corriste"))
````

Stemmer en ingl茅s

*Cuidado: las reglas de stemmer son dependientes del idioma*

````
from nltk.stem.snowball import EnglishStemmer

# Crear una instancia del SnowballStemmer para ingl茅s
stemmer = EnglishStemmer()

# Aplicar stemming a diferentes formas de la palabra "correr"
print(stemmer.stem("corriendo")) # no se reduce, ya que no es una palabra inglesa
print(stemmer.stem("correr"))
print(stemmer.stem("corriste"))
````

Usando otro stemmer (pueden encontrar otros en https://www.nltk.org/api/nltk.stem.html)

````
from nltk.stem import PorterStemmer

# Crear una instancia del PorterStemmer
stemmer = PorterStemmer()

# Aplicar stemming a diferentes formas de la palabra "correr"
print(stemmer.stem("corriendo"))  # nuevamente no se reduce
print(stemmer.stem("correr"))
print(stemmer.stem("corriste"))
````

````
# lemmatization
from nltk.stem import WordNetLemmatizer

# Crear una instancia del WordNetLemmatizer
wnl = WordNetLemmatizer()

# Lematizar algunas palabras
print(wnl.lemmatize("whale"))   # no se reduce, ya que ya es la forma base)
print(wnl.lemmatize("whales"))  # se reduce a la forma base "whale"
print(wnl.lemmatize("whaling")) # no se reduce, ya que no se encuentra en WordNet
````
