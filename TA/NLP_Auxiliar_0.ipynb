### Creación básica de arreglos

Creando e imprimiendo arreglos de numpy (ndarray):
- las matrices unidimensionales se imprimen luego como filas,
- las bidimensionales como matrices y tridimensionales como listas de matrices.

````
# Crear un arreglo a partir de una lista de listas
print("Arreglo a partir de una lista de listas:")
print(np.array([[1., 0., 0.], [0., 1., 2.]]), "\n")

# Crear un arreglo de ceros
print("Arreglo de ceros:")
print(np.zeros(10), "\n")

# Crear un arreglo de ceros multidimensional
a = np.zeros((2, 3, 4))
print("Arreglo multidimensional de ceros:")
print(a, "\n")

# Crear un arreglo de unos
print("Arreglo de unos:")
print(np.ones(10), "\n")

# Crear un arreglo de unos multidimensional con tipo de datos especificado
b = np.ones((2, 3, 4), dtype='int')
print("Arreglo multidimensional de unos:")
print(b, "\n")

# Crear un arreglo de valores aleatorios entre 0 y 1
print("Arreglo de valores aleatorios:")
print(np.random.rand(5), "\n")

# Crear un arreglo multidimensional de valores aleatorios entre 0 y 1
c = np.random.rand(2, 3, 4)
print("Arreglo multidimensional de valores aleatorios:")
print(c, "\n")
````

Los atributos más importantes de un objeto ndarray son:

````
# Número de dimensiones del arreglo
print("Número de dimensiones del arreglo:", c.ndim)

# Forma del arreglo (tamaño en cada dimensión)
print("Forma del arreglo (tamaño en cada dimensión):", c.shape)

# Número total de elementos en el arreglo
print("Número total de elementos en el arreglo:", c.size)

# Tipo de datos de los elementos del arreglo
print("Tipo de datos de los elementos del arreglo:", c.dtype)
````

Creando arreglos a partir de sequencias de numeros

````
# Crear un arreglo de enteros en un rango especificado
print("Arreglo de enteros en un rango especificado:")
print(np.arange(10), "\n")

# Crear un arreglo de enteros en un rango especificado con un paso determinado
print("Arreglo de enteros en un rango especificado con un paso determinado:")
print(np.arange(3, 20, 2), "\n")

# Crear un arreglo con 5 números equidistantes entre 1 y 2
print("Arreglo con 5 números entre 1 y 2:")
print(np.linspace(1, 2, 5), "\n")
````

````
# Cambiando las dimensiones de un array (ravel, reshape, resize)

# Utilizar 'ravel' para aplanar el arreglo a una dimensión
d = c.ravel()
print("Arreglo aplanado:", d, "\n")
print("Forma del arreglo aplanado:", d.shape, "\n")
print("Forma del arreglo original:", c.shape, "\n")

# Utilizar 'reshape' para crear un nuevo arreglo con otra distribución de las dimensiones (no modifica el original)
d = c.reshape((2, 12))
print("Nuevo arreglo con otra distribución de las dimensiones:", d, "\n")
print("Forma del nuevo arreglo:", d.shape, "\n")

# Utilizar 'T' para crear un nuevo arreglo con la transpuesta
e = d.T
print("Nuevo arreglo con la transpuesta:", e, "\n")
print("Forma del nuevo arreglo con la transpuesta:", e.shape, "\n")

# Utilizar 'resize' para cambiar las dimensiones en el propio arreglo (modifica el original)
x = d.resize(2, 12)
print("Forma del arreglo modificado:", d.shape, "\n")
print("Valor devuelto por resize (None indica que el arreglo fue modificado in-place):", x)
````

````
# Crear un arreglo de unos con dimensiones similares a otro arreglo
b = np.ones_like(a)
print("Arreglo de unos con dimensiones similares a 'a':")
print(b, "\n")
print("Forma del arreglo 'b':", b.shape, "\n")

# Crear un arreglo vacío con dimensiones similares a otro arreglo
c = np.empty(a.shape)
print("Arreglo vacío con dimensiones similares a 'a':")
print(c, "\n")
print("Forma del arreglo 'c':", c.shape, "\n")
````

### Operaciones básicas sobre arreglos

````
# Crear algunos arreglos con lo que hemos aprendido
a = np.arange(22, 32)
b = np.arange(11, 21)
print("Forma del arreglo 'a':", a.shape)
print("Forma del arreglo 'b':", b.shape)
````

````
# Suma y resta de arreglos
print("Suma de arreglos 'a' y 'b':", a + b)
print("Resta de arreglos 'a' y 'b':", a - b)

# Producto punto a punto (elemento a elemento) de arreglos
print("Producto punto a punto de arreglos 'a' y 'b':", a * b)
````

````
# Cambiar la forma del arreglo 'a' a 2x5 y del arreglo 'b' a 5x2
a.resize(2, 5)
b.resize(5, 2)

print("Arreglo 'a' con nueva forma (2x5):")
print(a)
print("\nArreglo 'b' con nueva forma (5x2):")
print(b)
````

````
# Producto matricial entre los arreglos 'a' y 'b'
print("Producto matricial entre los arreglos 'a' y 'b':")
print(a @ b)

# Otro producto matricial entre los arreglos 'a' y 'b'
print("Otro producto matricial entre los arreglos 'a' y 'b':")
print(a.dot(b))
````

````
# Otras operaciones con los arreglos
print("Cuadrado de los elementos del arreglo 'a':")
print(a**2)

print("Parte entera de la raíz cuadrada de los elementos del arreglo 'a':")
print(np.floor(np.sqrt(a)))

print("Pi elevado a la potencia de cada elemento del arreglo 'b' menos 10:")
print(np.pi**(b-10))
````

````
# Cambiar la forma del arreglo 'b' a 2x5 y asignarlo a 'e'
e = b.reshape(2, 5)
print("Arreglo 'e' con forma (2x5):")
print(e)

# Al especificar el parámetro axis, se puede aplicar una operación a lo largo de la dimensión especificada
# En este caso, se encuentra el máximo valor a lo largo del eje 1 (a lo largo de las filas) en el arreglo 'e'
print("Máximo valor a lo largo de las filas en el arreglo 'e':")
print(e.max(axis=1))
````

````
# Cambiar un valor específico en el arreglo 'e'
e[0, 0] = 1000
print("Arreglo 'e' después de cambiar el valor en la posición (0, 0):")
print(e)
````

### Broadcasting

Muchas operaciones en NumPy soportan [Broadcasting Semantics](https://docs.scipy.org/doc/numpy/user/basics.broadcasting.html#module-numpy.doc.broadcasting).

Si una operación NumPy admite broadcast, sus argumentos (ndarrays) se pueden expandir automáticamente para que sean del mismo tamaño (**sin hacer copias de los datos**). *No es más que una forma de hacer operaciones punto a punto sobre ndarrays que no necesariamente tienen las mismas dimensiones.*

Dos `ndarray` son "broadcastable" si:
1. Cada array tiene al menos una dimensión.
2. Al iterar sobre los tamaños de las dimensiones (comenzando en la dimensión final), sus tamaños deben ser iguales, uno de ellos es `1` o uno de ellos no existe.

Si dos `ndarray` son "broadcastable" las dimensiones del `ndarray` resultado se determina mediante:
1. Si el número de dimensiones no es igual, se antepone `1` a las dimensiones del `ndarray` con menos dimensiones para que tengan la misma longitud.
2. Luego, para cada tamaño de dimensión, el tamaño de dimensión resultante es el máximo de los tamaños en esa dimensión.

````
# Definir los arreglos con np.random.randint
a = np.random.randint(0, 10, size=(2, 1))
b = np.random.randint(0, 10, size=(1, 8))

# Mostrar los arreglos
print("Arreglo 'a':")
print(a)
print("\nArreglo 'b':")
print(b)

# Realizar la suma y obtener la forma resultante
result = a + b

# Mostrar el resultado de la suma y su forma
print("\nResultado de la suma:")
print(result)
print("\nForma del resultado de la suma:", result.shape)
````

El broadcasting se aplica para extender automáticamente las dimensiones de ambos arreglos de manera que sean compatibles para la operación de suma. NumPy agrega dimensiones adicionales con tamaño 1 a los arreglos que lo necesiten, permitiendo así la realización de operaciones elemento a elemento. En este caso, el resultado es un arreglo con forma (2, 8), donde cada elemento es la suma del correspondiente elemento en 'a' y 'b'.

### Tokenization

Importar el tokenizador de palabras recomendado por NLTK (actualmente ```nltk.tokenize.treebank.TreebankWordTokenizer``` junto con ```nltk.tokenize.punkt.PunktSentenceTokenizer``` para el idioma especificado)


````
# Importar tokenizador
from nltk import word_tokenize

# Frase de ejemplo
s = "Hola como están, esta es una frase en #español :)"

# Tokenizar la frase en español
tokens = word_tokenize(s, language="spanish")

# Mostrar los tokens resultantes
tokens
````

Importar otro tokenizer (pueden encontrar otros en https://www.nltk.org/api/nltk.tokenize.html)

````
# Otro tokenizador: tokenizador de tweets
from nltk.tokenize.casual import TweetTokenizer

# Crear una instancia del tokenizador
tknzr = TweetTokenizer()

# Tokenizar la frase utilizando el tokenizador
tokens = tknzr.tokenize(s)

# Mostrar los tokens resultantes
tokens
````

### Stemming y Lemmatization

Importar stemmer ```SnowballStemmer``` desde NLTK. SnowballStemmer soporta los lenguajes: Arabic, Danish, Dutch, English, Finnish, French, German, Hungarian, Italian, Norwegian, Portuguese, Romanian, Russian, Spanish and Swedish.

````
# Importar stemmer
from nltk.stem import SnowballStemmer

# Crear una instancia del SnowballStemmer para español
stemmer = SnowballStemmer("spanish")

# Aplicar stemming a diferentes formas de la palabra "correr"
print(stemmer.stem("corriendo"))
print(stemmer.stem("correr"))
print(stemmer.stem("corriste"))
````

Stemmer en inglés

*Cuidado: las reglas de stemmer son dependientes del idioma*

````
from nltk.stem.snowball import EnglishStemmer

# Crear una instancia del SnowballStemmer para inglés
stemmer = EnglishStemmer()

# Aplicar stemming a diferentes formas de la palabra "correr"
print(stemmer.stem("corriendo")) # no se reduce, ya que no es una palabra inglesa
print(stemmer.stem("correr"))
print(stemmer.stem("corriste"))
````

Usando otro stemmer (pueden encontrar otros en https://www.nltk.org/api/nltk.stem.html)

````
from nltk.stem import PorterStemmer

# Crear una instancia del PorterStemmer
stemmer = PorterStemmer()

# Aplicar stemming a diferentes formas de la palabra "correr"
print(stemmer.stem("corriendo"))  # nuevamente no se reduce
print(stemmer.stem("correr"))
print(stemmer.stem("corriste"))
````

````
# lemmatization
from nltk.stem import WordNetLemmatizer

# Crear una instancia del WordNetLemmatizer
wnl = WordNetLemmatizer()

# Lematizar algunas palabras
print(wnl.lemmatize("whale"))   # no se reduce, ya que ya es la forma base)
print(wnl.lemmatize("whales"))  # se reduce a la forma base "whale"
print(wnl.lemmatize("whaling")) # no se reduce, ya que no se encuentra en WordNet
````
